```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

#設定
L_axis = 25                #座標軸的範圍
spacing = 10              #座標間距

#建立座標軸
x = np.linspace(-L_axis,L_axis,spacing)
y = np.linspace(-L_axis,L_axis,spacing)
z = np.linspace(-L_axis,L_axis,spacing)

x,y,z = np.meshgrid(x,y,z)
fig = plt.figure()
ax = fig.gca(projection='3d')

X=np.array([-5,2,3])
v=np.array([0,3,0])
M=1
Q=1
Y1=[]
Y2=[]
Y3=[]
E=np.array([3,6,100])
#ax.quiver(x,y,z,E[0],E[1],E[2],color='g',length=2,normalize=True)

for t in range(0,50):
    def B(x,y,z):
        B=np.array([x*y,0,0])
        return B
    B=B(X[0],X[1],X[2])
    #ax.quiver(x,y,z,B[0],B[1],B[2],color='b',length=2,normalize=True)
    F=Q*E+Q*(np.cross(v,B))
    a=F/M
    v=v+a*0.01
    X=X+v*0.01
    Y1.append(X[0])
    Y2.append(X[1])
    Y3.append(X[2])
    
ax.scatter(Y1,Y2,Y3,marker='o', color="red")     
plt.show()
```
