```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

# x,y,z(-4410)
x = np.linspace(-4,4,10)
y = np.linspace(-4,4,10)
z = np.linspace(-4,4,10)

# 3D
x,y,z = np.meshgrid(x,y,z)

# 3D
fig = plt.figure(figsize=(12,12))
ax = fig.gca(projection='3d')

# (m,n),(x-m,y-n)
def B(x,y,m,n):
    i = 1                                           #Amps in the wire
    mu = 1.26 * 10**(-6)                            #Magnetic constant    (H/m)                   
    mag = (mu/(2*np.pi))*(i/np.sqrt((x-m)**2+(y-n)**2)) #Magnitude of the vector B 
    by = mag * (np.cos(np.arctan2(y-n,x-m)))            #By y arctan2(y,x)tan(x/y)
    bx = mag * (-np.sin(np.arctan2(y-n,x-m)))           #Bx x
    bz = z*0                                        #Bz (zero, using the right-hand rule)
    return bx,by,bz

def cylinder(r,m,n):   #m,n
    phi = np.linspace(-np.pi,np.pi,100)
    x = m+r*np.cos(phi) # +m
    y = n+r*np.sin(phi) # +n
    return x,y

# 
Bx=0
By=0
Bz=0

# 3*3,
n = 3
position = np.zeros([n, n]) # ([row, col])  #,(1,1)(3,3)9
for i in range(0,n):
    for j in range(0,n):
        position[i][j]= input("enter:")
print(position) #

#
for i in range(0,n):
    for j in range(0,n):
        if position[i][j]==0: # 0,
            continue
        bx,by,bz = B(x,y,i,j)           #Magnetic field
        Bx=(Bx+bx*position[i][j])*50                        #+bx
        By=(By+by*position[i][j])*50                        #+by    
        Bz=Bz+bz                        #+bz
        cx,cy = cylinder(0.2,i,j)       #Wire
        for l in np.linspace(-4,4,800):  #-44,8    #Plot the wire
            ax.plot(cx,cy,l,label='Cylinder',color='r') 
            #ax.plot(x,y,z)  -44800
                        
ax.quiver(x,y,z,Bx,By,Bz,color='b',normalize=False)  # ...
plt.title('Magnetic field of a straight wire')
plt.xlabel('x')
plt.ylabel('y')
ax.view_init(elev=80, azim=-30)
```
